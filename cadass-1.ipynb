```julia
using Plots

f(x,y) = 200 - x^2 - 2y^2
xs = range(-10*(2^0.5), 10*(2^0.5),length =1000)
ys = range(-10, 10,length=1000)


z = [max(f(xi, yi),0) for yi in ys, xi in xs]

surface(xs, ys, z,
        xlabel="x", ylabel="y", zlabel="f(x,y)",
        title="3D Surface plot", colorbar_title ="f(x,y)")
```

```julia
using Plots
using ForwardDiff

f(x, y) = 200 - x^2 - 2*y^2

f_x(x, y) = ForwardDiff.derivative(x_ -> f(x_, y), x)
f_y(x, y) = ForwardDiff.derivative(y_ -> f(x, y_), y)

F(x, y) = (f_x(x, y), f_y(x, y))

x1 = -1:1:1
y1 = -1:1:1

X = repeat(x1, inner=length(y1))
Y = repeat(y1, outer=length(x1))
U = [F(xi, yi)[1] for (xi, yi) in zip(X, Y)]
V = [F(xi, yi)[2] for (xi, yi) in zip(X, Y)]

quiver(X, Y, quiver=(U, V), xlabel="x", ylabel="y", color=:violet,
       title="Gradient Vector Field  of f(x,y) = 200 - x^2 - 2y^2")
```

```julia
using CalculusWithJulia, Plots

f(u) = [exp(u[1]) * u[2]^2, u[1] + 2u[2]]

c = curl(f)

x = -2:0.5:2
y = -2:0.5:2

X = repeat(x, inner=length(y))
Y = repeat(y, outer=length(x))

Zf(x,y) = c([x,y])
Z= [Zf(xi, yi) for (xi, yi) in zip(X,Y)]

Z_x(x, y) = ForwardDiff.derivative(x_ -> Zf(x_, y), x)
Z_y(x, y) = ForwardDiff.derivative(y_ -> Zf(x, y_), y)

normal = [[Z_x(xi, yi), Z_y(xi, yi), 0.0] for (xi, yi) in zip(X,Y) ]

U = [n[1] for n in normal]
V = [n[2] for n in normal]
W = [n[3] for n in normal]

L = [sqrt(u^2+v^2+w^2) for (u,v,w) in zip(U,V,W)]
U = U ./ L
V = V ./ L
W = W ./ L

surface(X, Y, Z; xlabel="x", ylabel="y",
    title="Curl of f(x,y) = (ey)e + (x+2y)e",  colorbar_title="curl")
quiver!(X,Y,Z, quiver=(U,V,W),color=:red)

```

```julia
using Plots
q = 2
l = 10
L = 1.25*l
Ay = 0.375*q*L
By = 0.625*q*L

function SfdBmd(q,l,L,Ay,By)
    x = 0:0.1:L
    Sf = zeros(length(x))
    for (i, xi) in enumerate(x)
        if xi<=l 
            Sf[i] = Ay - (q*xi)
        else
            Sf[i] = Ay + By - (q*xi)
        end
    end

    Bm = zeros(length(x))
    for (i, xi) in enumerate(x)
        if xi<=l 
            Bm[i] = (Ay*xi) - (q*xi*xi*0.5)
        else
            Bm[i] = (Ay*xi) + By*(xi-l) - (q*xi*xi*0.5)
        end
    end
        
    p1 = plot(x,Sf, xlabel = "distance from hinge (m)", ylabel = "Shear Force (N)",
              title = "SFD", legend = false, color=:green)
    p1 = plot!([0,0], [0,Sf[1]],color=:green)
     
    p1 = plot!([0,L], [0,0], color=:purple)

    p2 = plot(x,Bm, xlabel = "distance from hinge (m)", ylabel = "Bending Moment (Nm)", 
              title = "BMD", legend = false, color=:green)
    p2 = plot!([0,0], [0,Bm[1]],color=:green)
    p2 = plot!([L,L], [0,Bm[length(x)]],color=:green)
    p2 = plot!([0,L], [0,0])

    plot(p1,p2 ; layout = (2,1))
end

SfdBmd(q,l,L,Ay,By)
```

```julia
using Plots
q = 2
l = 10
L = 2*l
Ay = 0.4*q*l
By = 0.98*q*l
Cy = 0.42*q*l

function SfdBmd(q,l,L,Ay,By)
    x = 0:0.01:L
    Sf = zeros(length(x))
    for (i, xi) in enumerate(x)
        if xi<=0.4l
            Sf[i] = Ay
        elseif xi>0.4l && xi<=l
            Sf[i] = Ay - 0.8*q*l
        else
            Sf[i] = Ay + By - 0.8*q*l - q*(xi-l)
        end
    end

    Bm = zeros(length(x))
    for (i, xi) in enumerate(x)
        if xi<=0.4l
            Bm[i] = (Ay*xi)
        elseif xi>0.4l && xi<=l
            Bm[i] = (Ay*xi) - (0.8*q*l*(xi-0.4*l))
        else
            Bm[i] = (Ay*xi)+(By*(xi-l))-(0.8*q*l*(xi-0.4*l))-(q*(xi-l)*(xi-l)*0.5)
        end
    end
        
    p1 = plot(x,Sf, xlabel = "distance from hinge (m)", ylabel = "Shear Force (N)",
              title = "SFD", legend = false, color=:green)
    p1 = plot!([0,0], [0,Sf[1]],color=:green)
    p1 = plot!([L,L], [0,Sf[length(x)]],color=:green)
    p1 = plot!([0,L], [0,0])

    p2 = plot(x,Bm, xlabel = "distance from hinge (m)", ylabel = "Bending Moment (N)",
              title = "BMD", legend = false, color=:green)
    p2 = plot!([0,0], [0,Bm[1]],color=:green)
    p2 = plot!([L,L], [0,Bm[length(x)]],color=:green)
    p2 = plot!([0,L], [0,0])

    plot(p1,p2 ; layout = (2,1))
  
end

SfdBmd(q,l,L,Ay,By)
```
